```python
import pandas as pd
```

```python
df = pd.read_csv("survey_data_updated.csv")
```

```python
#LanguageHaveWorkedWith

language_usage = (
    df["LanguageHaveWorkedWith"]
    .dropna()
    .str.split(";")
    .explode()
    .str.strip()
)

language_df = pd.DataFrame({
    "LanguageHaveWorkedWith": language_usage
})

language_df.to_csv("LanguageHaveWorkedWith.csv", index=False)
```

```python
#DatabaseHaveWorkedWith

database_usage = (
     df["DatabaseHaveWorkedWith"]
    .dropna()
    .str.split(";")
    .explode()
    .str.strip()
)

database_df = pd.DataFrame({
    "DatabaseHaveWorkedWith": database_usage
})

database_df = database_df.reset_index(drop=True)

database_df.head()

database_df.to_csv("DatabaseHaveWorkedWith.csv", index=False)
```

```python
#PlatformHaveWorkedWith

platform_usage = (
    df["PlatformHaveWorkedWith"]
    .dropna()
    .str.split(";")
    .explode()
    .str.strip()
)

platform_df = platform_usage.reset_index(drop = True)
platform_df.head(10)

platform_df.to_csv("PlatformHaveWorkedWith.csv", index=False)
```

```python
#WebFrameHaveWorkedWith

df["WebframeHaveWorkedWith"]

webframe_usage = (
    df["WebframeHaveWorkedWith"]
    .dropna()
    .str.split(";")
    .explode()
    .str.strip()
)

webframe_df = webframe_usage.reset_index(drop = True)
webframe_df.head(10)
webframe_df.to_csv("WebframeHaveWorkedWith.csv", index=False)
```
